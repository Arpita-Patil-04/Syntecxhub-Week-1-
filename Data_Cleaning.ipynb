```python
import pandas as pd

data = {
    "Name": ["Arpita", "Rahul", "Sneha", "Amit", "Arpita", "Riya", "Karan"],
    " Age ": ["21", "", "22", "21", "21", "twenty", "23"],
    "Email": ["arpita@gmail.com", "rahul@gmail.com", "", "amit@gmail.com", "arpita@gmail.com", "riya@gmail.com", "karan@gmail.com"],
    "Join Date": ["2023-01-15", "15/02/2023", "2023/03/10", "", "2023-01-15", "2023-05-01", "01-06-2023"],
    "Salary": ["30000", "35000", "", "28000", "30000", "40000", ""],
    "Department": ["IT", "HR", "Finance", "IT", "IT", "HR", ""]
}

df = pd.DataFrame(data)
df.to_csv("raw_data.csv", index=False)

print("✅ Dataset created as raw_data.csv")
df.head()
```

```python
def clean_data(input_file, output_file):
    import pandas as pd

    log = []

    df = pd.read_csv(input_file)
    log.append(f"Initial shape: {df.shape}")

    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    log.append("Column names standardized.")

    before = len(df)
    df.drop_duplicates(inplace=True)
    log.append(f"Removed {before - len(df)} duplicate rows.")

    df['age'] = pd.to_numeric(df['age'], errors='coerce')

    df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

    df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')

    log.append("Converted age, salary, and join_date to correct data types.")

    missing_before = df.isnull().sum().sum()

    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in num_cols:
        df[col] = df[col].fillna(df[col].median())

    cat_cols = df.select_dtypes(include=['object']).columns
    for col in cat_cols:
        if not df[col].mode().empty:
            df[col] = df[col].fillna(df[col].mode()[0])

    missing_after = df.isnull().sum().sum()
    log.append(f"Handled {missing_before - missing_after} missing values.")

    df.to_csv(output_file, index=False)
    log.append(f"Saved cleaned dataset as {output_file}")

    with open("cleaning_log.txt", "w") as f:
        for entry in log:
            f.write(entry + "\n")

    print("✅ Data Cleaning Completed!\n")
    print("---- CLEANING LOG ----")
    for entry in log:
        print(entry)
```

```python
clean_data("raw_data.csv", "cleaned_data.csv")
```
